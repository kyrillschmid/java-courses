```java
import java.util.Random;
import java.util.Scanner;

public class Game {
    private static final int MAX_HEALTH = 10;

    private static final int KNIGHT = 1;
    private static final int WIZARD = 2;
    private static final int ARCHER = 3;

    private static final int ATTACK = 1;
    private static final int DEFEND = 2;
    private static final int HEAL = 3;

    private static final Scanner scanner = new Scanner(System.in);
    private static final Random random = new Random();

    public static void play() {
        int playerChoice = getPlayerChoice();
        int computerChoice = getComputerChoice();

        int playerHealth = MAX_HEALTH;
        int computerHealth = MAX_HEALTH;

        while (playerHealth > 0 && computerHealth > 0) {
            int playerAction = getPlayerAction();
            int computerAction = getComputerAction(computerChoice);

            if (playerAction == ATTACK) {
                switch (playerChoice) {
                    case KNIGHT:
                        int knightDamage = random.nextInt(10) + 1;
                        computerHealth -= knightDamage;
                        System.out.println("You attacked with your sword for " + knightDamage + " damage!");
                        break;
                    case WIZARD:
                        int wizardDamage = random.nextInt(20) + 1;
                        computerHealth -= wizardDamage;
                        System.out.println("You attacked with your magic for " + wizardDamage + " damage!");
                        break;
                    case ARCHER:
                        int archerDamage = random.nextInt(15) + 1;
                        computerHealth -= archerDamage;
                        System.out.println("You attacked with your bow for " + archerDamage + " damage!");
                        break;
                }
            } else if (playerAction == DEFEND) {
                int damage = random.nextInt(5) + 1;
                playerHealth -= damage;
                System.out.println("You defended and took " + damage + " damage!");
            } else if (playerAction == HEAL) {
                int healAmount = random.nextInt(15) + 1;
                playerHealth += healAmount;
                System.out.println("You healed for " + healAmount + " health!");
            }

            if (computerAction == ATTACK) {
                switch (computerChoice) {
                    case KNIGHT:
                        int knightDamage = random.nextInt(10) + 1;
                        playerHealth -= knightDamage;
                        System.out.println("The knight attacked you with its sword for " + knightDamage + " damage!");
                        break;
                    case WIZARD:
                        int wizardDamage = random.nextInt(20) + 1;
                        playerHealth -= wizardDamage;
                        System.out.println("The wizard attacked you with its magic for " + wizardDamage + " damage!");
                        break;
                    case ARCHER:
                        int archerDamage = random.nextInt(15) + 1;
                        playerHealth -= archerDamage;
                        System.out.println("The archer attacked you with its bow for " + archerDamage + " damage!");
                        break;
                }
            } else if (computerAction == DEFEND) {
                int damage = random.nextInt(5) + 1;
                computerHealth -= damage;
                System.out.println("The computer defended and took " + damage + " damage!");
            } else if (computerAction == HEAL) {
                int healAmount = random.nextInt(15) + 1;
                computerHealth += healAmount;
                System.out.println("The computer healed for " + healAmount + " health!");
            }

                
        }

        if (playerHealth <= 0) {
            System.out.println("You lose!");
        } else {
            System.out.println("You win!");
        }
    }

    private static int getPlayerChoice() {
        System.out.println("Choose your avatar:");
        System.out.println("1. Knight");
        System.out.println("2. Wizard");
        System.out.println("3. Archer");
        int choice = scanner.nextInt();
        while (choice < 1 || choice > 3) {
            System.out.println("Invalid choice, please choose again.");
            choice = scanner.nextInt();
        }
        return choice;
    }

    private static int getComputerChoice() {
        return random.nextInt(3) + 1;
    }

    private static int getPlayerAction() {
        System.out.println("Choose your action:");
        System.out.println("1. Attack");
        System.out.println("2. Defend");
        System.out.println("3. Heal");
        int action = scanner.nextInt();
        while (action < 1 || action > 3) {
            System.out.println("Invalid action, please choose again.");
            action = scanner.nextInt();
        }
        return action;
    }

    private static int getComputerAction(int computerChoice) {
        return random.nextInt(3) + 1;
    }
}

```

```java
Game g = new Game();
```
